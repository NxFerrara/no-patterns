```python
from open3d import *
import scipy.io
import pandas as pd
import numpy as np
import glob
import os
```

```python
def load_file(filedir):
    mat = scipy.io.loadmat(filedir)
    try:
        points = mat["points"]
    except KeyError as e:
        print(f'Could not load {file} file -> KeyError: {e}')
        return []
    return points

def read_pcd(file):
    return read_point_cloud(file)

def write_pcd(filedir, points):
    pcd = PointCloud()
    pcd.points = Vector3dVector(points)
    write_point_cloud(filedir, pcd)

def mat_to_pcd(filedir, extension, newdir):
    files = glob.glob(filedir+"/*"+extension)
    if not os.path.exists(newdir):
        os.mkdir(newdir)
        print(newdir+" directory created")
    for file in files:
        points = load_file(file)
        if len(points) != 0:
            filename = file.split(".")[0]+".ply"
            filename = filename.split("/")[-1]
            filename = newdir+filename
            if not os.path.exists(filename):
                write_pcd(filename, points)


def create_model(filedir):
    data = []
    with open(filedir) as f:
        file = f.read().splitlines()
        for item in file:
            item = item.split(' ')
            new_item = []
            for e in item:
                if e != '':
                    new_item.append(e)
            data.append(new_item)
    vertices = []
    triangles = []
    for i in range(int(data[0][0])):
        vertices.append(list(map(float, data[i+1][:3])))

    for i in range(int(data[0][1])):
        triangles.append(list(map(int, data[i+int(data[0][0])+1])))
        
    mesh = open3d.geometry.TriangleMesh()
    mesh.vertices = Vector3dVector(np.array(vertices))
    mesh.triangles = Vector3iVector(np.array(triangles))
    mesh.compute_triangle_normals()
    mesh.compute_vertex_normals()
    return mesh
```
